```python
from sklearn.impute import SimpleImputer

mean_columns = ["id_02", "id_06", "id_11", "id_13", 'id_19', 'id_20', 'card3', 'card5']
median_columns = ["id_32", 'addr2', 'V138', 'V139', 'V140', 'V141', 'V142', 'V143', 'V144', 'V145', 'V146', 'V147', 'V148', 'V149']
mode_columns = [
    "id_05", "id_14", "id_17", 'card2', "V152", 'V161', 'V158', 'V162', 'V163', 'V165', 'V167', 'V168', 'V170', 'V171', 'V172', 
    'V173', 'V174', 'V175', 'V176', 'V177', 'V178', 'V179', 'V180', 'V181', 'V182', 'V183', 'V184', 'V185', 'V186', 'V187', 
    'V188', 'V189', 'V190', 'V191', 'V194', 'V195', 'V196', 'V197', 'V198', 'V199', 'V200', 'V201', 'V202', 'V203','V204','V205', 
    'V206', 'V207', 'V208', 'V209', 'V210','V211','V212','V213','V214','V215','V216','V217','V218','V219','V220','V221','V222',
    'V223','V224','V225','V226','V227','V228','V229','V230','V231','V232','V233','V234','V235','V236','V237','V238','V239','V240',
    'V241','V242','V243','V244','V245','V246','V247','V248','V249','V250','V251','V252','V253','V254','V255','V256','V257','V258',
    'V259','V260','V261','V262','V263','V264','V265','V266','V267','V268','V269','V270','V271', 'V272','V273','V274','V275','V276',
    'V277','V278','V279','V280','V281','V282','V283','V284','V285', 'V286', 'V287', 'V288', 'V289', 'V290', 'V291','V292','V293',
    'V294', 'V295','V296','V297','V298','V299','V300','V301','V302','V303','V304','V305','V306','V307','V308','V309','V310','V311',
    'V312','V313','V314','V315','V316','V317','V318','V319','V320','V321','V322','V323','V324','V325','V326','V327','V328','V329',
    'V330','V331','V332','V333','V334','V335', 'V336', 'V337', 'V338', 'V339', 'addr1'
]


# Imputers
imp_mean = SimpleImputer(strategy='mean')
imp_median = SimpleImputer(strategy='median')
imp_mode = SimpleImputer(strategy='most_frequent')

# Impute the columns
train[mean_columns] = imp_mean.fit_transform(train[mean_columns])
train[median_columns] = imp_median.fit_transform(train[median_columns])
train[mode_columns] = imp_mode.fit_transform(train[mode_columns])
test[mean_columns] = imp_mean.fit_transform(test[mean_columns])
test[median_columns] = imp_median.fit_transform(test[median_columns])
test[mode_columns] = imp_mode.fit_transform(test[mode_columns])



# fill the rest column with the standard deviation
# List of columns to fill with standard deviation

std_columns = ['V150', 'V151', 'V153', 'V154', 'V155', 'V156', 'V157', 'V159', 'V160', 'V164', 'V166', 'V169', 'V192', 'V193']


# Calculate standard deviations for the columns
std_values = train[std_columns].std().to_dict()

# Function to fill missing values with standard deviation
def fill_with_std(df, columns, std_values):
    for col in columns:
        df[col].fillna(std_values[col], inplace=True)
    return df

# Fill missing values in the train DataFrame
train = fill_with_std(train, std_columns, std_values)

# Fill missing values in the test DataFrame using the same standard deviation values
test = fill_with_std(test, std_columns, std_values)
```

```python
columns_with_nulls = [
    "id_15", "id_16", "id_28", "id_29", "id_30", "id_31", "id_33", "id_34", "id_35", 
    "id_36", "id_37", "id_38", "DeviceType", "DeviceInfo", "card4", "card6", 
    "P_emaildomain", "R_emaildomain"
]

train[columns_with_nulls].dtypes #data type of null values
```

```python
# Apply LabelEncoder to each column with nulls
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for column in columns_with_nulls:
    train[column] = encoder.fit_transform(train[column])
    test[column]=encoder.fit_transform(test[column])
```

```python
# Assuming columns_with_nulls contains the list of columns with null values

# Drop columns with nulls from train and test datasets
train_null = train.drop(columns_with_nulls, axis=1)
test_null = test.drop(columns_with_nulls, axis=1)

# Apply KNN imputer
from sklearn.impute import KNNImputer

# Initialize KNNImputer
kn = KNNImputer(n_neighbors=5)

# Fit and transform on train data
train_imputed = kn.fit_transform(train[columns_with_nulls])

# Transform test data (using the imputer learned from train data)
test_imputed = kn.transform(test[columns_with_nulls])

# Convert imputed arrays back to DataFrame
train_imputed_pd = pd.DataFrame(train_imputed, columns=columns_with_nulls)
test_imputed_pd = pd.DataFrame(test_imputed, columns=columns_with_nulls)

# Concatenate imputed data with non-null columns
train_final = pd.concat([train_null, train_imputed_pd], axis=1)
test_final = pd.concat([test_null, test_imputed_pd], axis=1)


```

```python
columns=['id_12', 'ProductCD']
for column in columns:
    train_final[column] = encoder.fit_transform(train_final[column])
    test_final[column]=encoder.fit_transform(test_final[column])
```

```python
# Get data ready for machine learning

colums_to_drop=["TransactionID", "isFraud", "TransactionDT" ]
X=train_final.drop(colums_to_drop,  axis=1)
y=train_final["isFraud"]# Import necessary libraries


```

```python
from sklearn.model_selection import TimeSeriesSplit
# Initialize the TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

# Iterate through the splits
for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Print or check the shapes of the splits
    print("X_train shape:", X_train.shape, "X_test shape:", X_test.shape)
    print("y_train shape:", y_train.shape, "y_test shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
Gr_model=RandomForestClassifier
model=Gr_model(n_estimators=100,random_state=0)
model.fit(X_train, y_train)
```

```python
# drop transaction id in test_final
drop=["TransactionID", "TransactionDT"]
test_predict=test_final.drop(drop, axis=1)
predictions=model.predict(test_predict)
```
